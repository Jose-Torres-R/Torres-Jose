```python
import pandas as pd
import numpy as np
import sklearn as sl
from statistics import mode
import matplotlib.pyplot as plt
```

Carga de base de datos

```python
df = pd.read_csv("datos/ejemplo_data.csv", encoding="utf-8", sep=",")
```

Identifique los tipos de varibles que hay disponibles en la base de datos (df.types o df.info()).

```python
print(df.head())
df.dtypes
df.info()
```

Utilizando la función astype transforme el atributo "ID" a entero y el atributo "Activo" a booleano.
Vuelva a consultar el estado de las variables.

```python
df["ID"] = df["ID"].astype("int64")
df.dtypes
```

Convierta el atributo "unidades" a entero y "2016" a flotante.

```python
df["Unidades"] = df["Unidades"].replace({"No":0}, regex=True)
df["Unidades"] = df["Unidades"].astype("int64")
df["2016"] = df["2016"].apply(lambda x: x.replace(",","").replace("$","").replace(".",""))
df["2016"] = df["2016"].astype("float64")
df.dtypes
```

Cargar la base de datos de nombre "ecommerce data.csv". En esta parte recomendamos explorar las diferentes opciones de read que tiene disponible la librería Pandas, identifícando los argumentos disponibles en cada una de ellas.

```python
df_ecod = pd.read_csv("datos/ecommerce_data.csv", encoding="latin-1", sep=",")
```

Identifíque los tipos de varibles que hay disponibles en la base de datos

```python
df_ecod.dtypes
df_ecod.info()
print(df_ecod.head())
```

Utilizando la función astype transforme el atributo "InvoiceNo" a entero y el atributo "Description" a string. Vuelva a consultar el estado de las variables.

```python
df_ecod["InvoiceNo"] = df_ecod["InvoiceNo"].apply(lambda x: x.replace("C","").replace("A",""))
df_ecod["InvoiceNo"] = df_ecod["InvoiceNo"].astype("int64")
df_ecod["Description"] = df_ecod["Description"].astype("string")
df_ecod.dtypes
```

Convierta el atributo "Quantity" a entero y "UnitPrice" a flotante.

```python
df_ecod["Quantity"] = df_ecod["Quantity"].astype("int64")
df_ecod["UnitPrice"] = df_ecod["UnitPrice"].astype("float64")
df_ecod.dtypes
```

La columna "InvoiceDate" contiene un string que representa "fecha-hora", separe la columna en dos columnas que representen cada atributo por separado.

```python
fechahora = df_ecod["InvoiceDate"].str.split(expand=True)
fechahora.columns = ["fecha", "hora"]
df_ecod = pd.concat([df_ecod, fechahora], axis=1)
df_ecod
```

Añada una nueva columna que represente el monto total para cada boleta

```python
df_ecod["total_boleta"] = df_ecod["Quantity"] * df_ecod["UnitPrice"]
df_ecod
```

Exporte la base de datos procesada en formato ".csv"

```python
df_ecod.to_csv("muestra.csv", index=False)
```
